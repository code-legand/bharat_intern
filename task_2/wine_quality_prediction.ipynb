### Transforming variables

```python
data['dissolved sulfur dioxide'] = data['total sulfur dioxide'] - data['free sulfur dioxide']
```

```python
data = data.drop(columns = ['total sulfur dioxide'])
```

```python
data.head()
```

### Removing outliers

```python
fig, ax = plt.subplots(3,4, figsize=(16, 10))
sns.boxplot(data, x= data['fixed acidity'], ax=ax[0, 0])
sns.boxplot(data, x= data['volatile acidity'], ax=ax[0, 1])
sns.boxplot(data, x= data['citric acid'], ax=ax[0, 2])
sns.boxplot(data, x= data['residual sugar'], ax=ax[0, 3])
sns.boxplot(data, x= data['chlorides'], ax=ax[1, 0])
sns.boxplot(data, x= data['free sulfur dioxide'], ax=ax[1, 1])
sns.boxplot(data, x= data['density'], ax=ax[1, 2])
sns.boxplot(data, x= data['pH'], ax=ax[1, 3])
sns.boxplot(data, x= data['sulphates'], ax=ax[2, 0])
sns.boxplot(data, x= data['alcohol'], ax=ax[2, 1])
sns.boxplot(data, x= data['dissolved sulfur dioxide'], ax=ax[2, 2])
plt.show()
```

```python
features = data.columns.tolist()
features.remove('quality')
for feature in features:
    q1 = data[feature].quantile(0.25)
    q3 = data[feature].quantile(0.75)
    iqr = q3-q1
    ul = q3 +1.5*iqr
    ll = q1 - 1.5*iqr
    
    data = data.loc[(data[feature]>ll) & (data[feature]<ul), :]
```

```python
data.info()
```

```python
data.describe()
```

### Balancing classes

```python
from sklearn.utils import resample
```

```python
data['quality'].value_counts()
```

```python
data_class_5 = data.loc[data['quality']==5, :]
data_class_6 = data.loc[data['quality']==6, :]
data_class_7 = data.loc[data['quality']==7, :]
```

```python
data_class_6 = resample(data_class_6, n_samples=332)
data_class_7 = resample(data_class_7, n_samples=332)
data = pd.concat((data_class_5, data_class_6, data_class_7))
```

```python
data.info()
```
